```python
import pandas as pd                            
bachelorette = pd.read_csv('bachelorette.csv')          #reads the csv file, organizes it into a table
```

```python
bachelorette = bachelorette.loc[bachelorette.SHOW == "Bachelorette"]    #selects the rows with Bachelorette (removed bachelor seasons)
print (bachelorette.head())
```

```python
bachelorette = bachelorette[bachelorette["SHOW"] != "SHOW"]
bachelorette = bachelorette[bachelorette["CONTESTANT"] != "ID"]     #source has multiple headers and need to clean/delete
print (bachelorette.head())
```

```python
bachelorette = bachelorette.replace([".", "", " "], pd.NA)          # changes missing placeholders to NaN
print(bachelorette.head())
```

```python
date_columns = [f"DATES-{i}" for i in range(1, 11)]   #list of the DATES column names
print(date_columns)
```

```python
bachelorette["first_1on1"] = False   # initialize column to False
```

```python
seasons = sorted(bachelorette["SEASON"].unique())     #loops through each season

for season in seasons:                                #loop goes through each season at a time
    print("Season", season)

    season_data = bachelorette[bachelorette["SEASON"] == season]    #uses data from this season only

    first_week = None
    week_number = 1   # initialize the week, start at 1

    for column in date_columns:
        if column not in season_data.columns:        # Skips columns that are missing for this season
            week_number +=1 
        else: 
            if (season_data[column] == "D1").any():   # check if any contestant got 1-on-1 that week
                first_week = week_number               # track the week number
                break                                  # stop because we found the first 1-on-1
            week_number = week_number + 1              # increase the counter

    if first_week is not None:
        first_column = f"DATES-{first_week}"
        first_names = season_data.loc[season_data[first_column] == "D1", "CONTESTANT"]        #showing the specific contestants who got a 1-on-1
        #print(first_names)
        bachelorette.loc[(bachelorette["SEASON"] == season) & (bachelorette["CONTESTANT"].isin(first_names)), "first_1on1"] = True   #puts contestants who got 1-on-1 for each season in the data frame, isin creates a boolean list, found on pandas.py.org
        print("First 1-on-1 contestant(s):", first_names.tolist())            #turns the series into a list, found on pandas.py.org
    else:
        print("No 1-on-1 found for this season")
```
