```python
bachelorette_rows = []
for line in open ('bachelorette.csv'): 
    cleaned_line = line.strip()                        #removes whitespace characters
    cleaned_list = cleaned_line.split(",")             #creates a list of strings for each line
    if cleaned_list[0] == 'Bachelorette':
       bachelorette_rows.append(cleaned_list)          #adding the cleaned list to bachelorette_rows
```

```python
import pandas as pd                            
bachelorette = pd.read_csv('bachelorette.csv')          #reads the csv file, organizes it into a table
```

```python
bachelorette = bachelorette.loc[bachelorette.SHOW == "Bachelorette"]    #selects the rows with Bachelorette (removed bachelor seasons)
print (bachelorette.head())
```

```python
bachelorette = bachelorette[bachelorette["SHOW"] != "SHOW"]
bachelorette = bachelorette[bachelorette["CONTESTANT"] != "ID"]     #source has multiple headers and need to clean/delete
print (bachelorette.head())
```

```python
bachelorette = bachelorette.replace([".", "", " "], pd.NA)          # changes missing placeholders to NaN
print(bachelorette.head())
```

```python
date_columns = [f"DATES-{i}" for i in range(1, 11)]   #list of the DATES column names
print(date_columns)
```

```python
bachelorette["first_1on1"] = False                        #initialize new column to False
bachelorette["top4"] = False                              #new column, initialized to false
bachelorette["made_top4"] = False              #intitialize new column to see if first 1-on-1 made top 4
```

```python
seasons = sorted(bachelorette["SEASON"].unique())     # loops through each season

for season in seasons:                                # loop goes through each season at a time
    print("Season", season, ":")
    print()                                          #creates a space

    season_data = bachelorette[bachelorette["SEASON"] == season]    # uses data from this season only

# Find the first week with the first 1-on-1
    
    first_week = None
    week_number = 1   # initialize the week, start at 1

    for column in date_columns:
        if column not in season_data.columns:        # Skips columns that are missing for this season
            week_number +=1 
        else: 
            if (season_data[column] == "D1").any():   # check if any contestant got 1-on-1 that week
                first_week = week_number               # track the week number
                break                                  # stop because we found the first 1-on-1
            week_number = week_number + 1              # increase the counter
            
# Find who got the 1-on-1
    
    if first_week is not None:
        first_column = f"DATES-{first_week}"
        first_names = season_data.loc[season_data[first_column] == "D1", "CONTESTANT"]        # showing the specific contestants who got a 1-on-1
        #print(first_names)
        bachelorette.loc[(bachelorette["SEASON"] == season) & (bachelorette["CONTESTANT"].isin(first_names)), "first_1on1"] = True   # puts contestants who got 1-on-1 for each season in the data frame, isin creates a boolean list, found on pandas.py.org
        print("First 1-on-1 contestant(s):", first_names.tolist())            # turns the series into a list, found on pandas.py.org
    else:
        print("No 1-on-1 found for this season")


    bachelorette.loc[bachelorette["SEASON"] == season, "date_count"] = bachelorette.loc[bachelorette["SEASON"] == season, date_columns].count(axis=1)  #Counts how many dates each contestant went on and create new column, axis=1 represents counts for each row

# Top 4 = contestants with the **highest** counts (not lowest eliminations)

    season_sorted = season_data.sort_values("date_count", ascending=False)     # sorts values from largest date_count to smallest, pandas.py.org 
    top4_names = season_sorted.head(4)["CONTESTANT"].tolist()            #turns the series into a list, selects top 4 contestants
    print("Top 4 contestants:", top4_names)

   
    bachelorette.loc[(bachelorette["SEASON"] == season) & (bachelorette["CONTESTANT"].isin(top4_names)), "top4"] = True   #shows who made top 4 

#Checks if first-1-on-1 contestants made Top 4
   
    made_it = [name for name in first_names if name in top4_names]        #checks to see who is in both lists of first_names and top4_names?
    #print(made_it)

    if len(made_it) > 0:
        print("FIRST 1-on-1 and made TOP 4:", made_it)
        print("--------------------")                                                          #creates a space
        bachelorette.loc[bachelorette["CONTESTANT"].isin(made_it),"made_top4"] = True         #commits to dataframe
    else:
        print("FIRST 1-on-1 but did not reach Top 4")
        print("--------------------")                                                          #creates a space
```

```python
#Data Summary for first 1-on-1 contestants who made Top 4

total_first_1on1 = bachelorette["first_1on1"].sum()                                                      # Total number of first 1-on-1 contestants across all seasons
first_1on1_made_top4 = bachelorette[bachelorette["first_1on1"] & bachelorette["made_top4"]].shape[0]     # Number of first 1-on-1 contestants who made Top 4
percent_contestants = (first_1on1_made_top4 / total_first_1on1) * 100                                    #Percentage of first 1-on-1 contestants who made Top 4
print(f"Percentage of first 1-on-1 contestants who made Top 4: {percent_contestants:.1f}%")              #.1f represents 1 digit after the decimal point, python.py.org
```

```python
total_seasons = len(bachelorette["SEASON"].unique())             # Counts how many seasons there are

# Count how many first-1-on-1 contestants made Top 4

seasons_with_first_1on1_top4 = []                                # Initialize the list of seasons where at least one first-1-on-1s made top 4

for season in bachelorette["SEASON"].unique():
    season_data = bachelorette[bachelorette["SEASON"] == season]
    if ((season_data["first_1on1"] == True) & (season_data["top4"] == True)).any():        # Check if any first-1-on-1s made top 4 in this season
        seasons_with_first_1on1_top4.append(season)


percent_seasons = len(seasons_with_first_1on1_top4) / total_seasons * 100                # Calculate percentage of seasons 
print("Seasons where the contestant with the first 1-on-1 made Top 4:", seasons_with_first_1on1_top4)
print(f"Percentage of seasons where first 1-on-1 made Top 4: {percent_seasons:.1f}%") 
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# Count the seasons where contestants had the first 1-on-1 and made Top 4
seasons = bachelorette["SEASON"].unique()

success_count = 0
for season in seasons:
    season_data = bachelorette[bachelorette["SEASON"] == season]
    if ((season_data["first_1on1"]) & (season_data["made_top4"])).any():
        success_count += 1

fail_count = len(seasons) - success_count

# Create Labels and values
labels = ["Made Top 4", "Did Not Make Top 4"]
values = [success_count, fail_count]

x = np.arange(len(values))  

# Plot
fig, ax = plt.subplots(figsize=(4,2))

ax.bar(x, values)

ax.set_xticks(x)
ax.set_xticklabels(labels)

ax.set_ylabel("Number of Seasons")

plt.title("How Often First 1-on-1 Contestant Makes Top 4")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.style.use('_mpl-gallery-nogrid')

labels = ["Made Top 4", "Did Not Make Top 4"]
sizes = [success_count, fail_count]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

plt.title("Percentage of First 1-on-1 Contestants Who Made Top 4")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.style.use('_mpl-gallery')

season_totals = (
    bachelorette.groupby("SEASON")[date_columns].count().sum(axis=1)
)

fig, ax = plt.subplots(figsize=(5, 2))   # Adds space at the bottom 

ax.plot(season_totals.index, season_totals.values, marker="o")

ax.set(
    xlabel="Season",
    ylabel="Total Number of Dates",
    title="Total Dates Per Season"
)

plt.show()
```
